Create a DataFrame based on the following data and set the column name to “Stamp”. With the apply function generate the “Year”, “Month”, “Day”, “Hour”, “Minute” and “Second” columns from the “Stamp” column.

```python
import pandas as pd 
# Create date and time with dataframe 
rng = pd.DataFrame() 
rng['stamp'] = pd.date_range('12/1/2016', periods = 72, freq ='3910S') 

# Print the dates in dd-mm-yy format 
rng[:5] 

# Create features for year, month, day, hour, and minute 
rng['year'] = rng['stamp'].dt.year 
rng['month'] = rng['stamp'].dt.month 
rng['day'] = rng['stamp'].dt.day 
rng['hour'] = rng['stamp'].dt.hour 
rng['minute'] = rng['stamp'].dt.minute 
rng['second'] = rng['stamp'].dt.second

# Print the dates divided into features 
rng.head(12) 
```
