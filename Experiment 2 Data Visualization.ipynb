### 3.1 Code

````python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
````

### 3.2 Prepare data

````python
data= pd.read_csv("spx.csv", index_col=0, parse_dates=True)
````

````python
data
````

````python
spx = data.SPX
spx = data.SPX
spx = spx['2007-01-01':'2010-12-31']
````

````python
crisis_data = [
(datetime(2007, 10, 11), 'Peak of bull market'),
(datetime(2008, 3, 12), 'Bear Stearns Fails'),
(datetime(2008, 9, 15), 'Lehman Bankruptcy')
]
````

### 3.3 Plotting

````python
fig = plt.figure(figsize = (25, 10))
ax1 = fig.add_subplot(1, 2, 1)
ax1.set_ylim([600, 1800])
#ax1.set_xlim(['01/01/2007', '01/01/2011'])

ax1.set_title("Important Dates in the 2008-2009 financial crisis", c = 'g',
fontsize = 20)
for date, label in crisis_data:
    ax1.annotate('Peak of bull market', xy = (date, spx.asof(date) + 50),
    xytext = (date, spx.asof(date) + 225),
        arrowprops = dict(fc='k', shrink=0.001, headwidth = 10.0,
    width = 3.0))
ax1.plot(spx)
ax2 = fig.add_subplot(1, 2, 2)
ax2.set_title('Histogram in the 2008-2010 Financial Crisis', c = 'g', fontsize
= 20)
ax2.hist(spx, bins = 80, orientation = 'horizontal')
ax2.plot()
````
